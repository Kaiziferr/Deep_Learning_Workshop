# **Resumen**

```python
summarization = pipeline(task='summarization')
```

```python
texto = """
Desde este miércoles 31 de mayo, el Valle del Cauca estrena una moderna Unidad Móvil de fauna silvestre, este vehículo está adaptado para el transporte seguro de animales silvestres, bien sea para llevarlos a su sitio de valoración y tratamiento o a su liberación.

“Estamos felices de entregar al servicio de la fauna (…) este moderno vehículo que brindará más seguridad y confort a los animales silvestres que por alguna razón deban ser transportados, ya sea porque los llevemos para ser atendidos por nuestros especialistas o porque felizmente pudieron ser rehabilitados y los llevamos de vuelta a su hábitat natural”, dijo durante la entrega Marco Antonio Suárez Gutiérrez, director general de la Corporación autónoma regional del Valle del Cauca (CVC).
"""
```

```python
summarization(texto, min_length=30)
```
